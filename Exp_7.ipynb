#***Study Of While Loop***

```python
# Print i as long as i is less than 6:
i = 1
while i < 6:
  print(i)
  i +=1
```

```python
# Print numbers from 1 to N
n = int(input("Enter N: "))
i = 1
while i <= n:
  print(i)
  i += 1
```

#***Factorial Of A Number***

```python
#Factorial of a number
n = int(input("Enter N: "))
fact = 1
while n > 0:
  fact *= n
  n -= 1
print("Factorial =" ,fact)
```

#***Fibonacci Series using while loop***


```python
#Fibonacci Series using while loop
n = int(input("Enter number of terms:"))
a = 0
b = 1
i = 1
while i <= n:
  print(a,end = " ")
  c = a + b
  a = b
  b = c
  i += 1
```

```python
limit = int(input("Enter limit: "))
a = 0
b = 1
while a <= limit:
    print(a, end=" ")
    a, b = b, a + b
```

#***Reverse Of A Number***

```python
#reverse of a number
n=int(input("Enter Number: "))
rev = 0
while n>0:
  digital = n%10
  rev=rev * 10+ digital
  n//= 10
print("Reversed Number: ",rev)
```

#***Palindrome Of a Number/String***

```python
#Check Palindrome Number / string
num= int(input("Enter Number: "))
temp = num
rev = 0
while num>0:
  rev = rev * 10 + num % 10
  num//=10
if temp ==rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
s= "madam"
i,j = 0, len(s)-1
is_pelindrome = True
while i < j:
  if s[i] != s[j]:
    is_pelindrome = False
    break
  i += 1
  j -= 1
if is_pelindrome:
  print("Pelindrome")
else:
  print("Not Pelindrome")
```

```python
st = input("Enter the String: ")
rev = st[::-1]
if st==rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

#***Count Digits In A Number***

```python
#count digits in a number
num = int(input("Enter Number: "))
count = 0
while num>0:
  count += 1
  num = num//10
print("Number of digits =", count)
```

#***Search In List***

```python
#search an element in a list
nums = [10,20,30,40,50]
keys = int(input("Enter Element ot search: "))

i = 0
while i<len(nums):
  if nums[i] == keys:
    print("Element found at index: ",i)
    break
  i+=1
else :
  print("element not found")
```
